```python
import pandas as pd

data = {"month":["October","April","May","June"],"age":[20,None,30,16],"job":["engineer","doctor","athletic","carpenter"]}
frame = pd.DataFrame(data)
# columns are placed according to the order of the keys in data
print(frame)
```

```python
# specifying column order
frame1 = pd.DataFrame(data, columns=['age','job','month'])
print(frame1)
```

```python
# passing a column not in the dict will result in NaN values for that column
frame2 = pd.DataFrame(data, columns=['age','job','month','salary'])
print(frame2)
print(frame2.columns)
```

```python
# get a column's data using frame.column_name or frame['column_name']
# frame['column_name'] is preferred since column names may conflict with built-in attributes
print(frame2.age)
print(frame2['age'])
```

```python
# get row data using loc[] with labeled index
labled = pd.DataFrame(data, index=["one","two","three","four"])
print(labled.loc["two"])
```

```python
# assigning a scalar value creates a column filled with that value
labled["debt"] = 12
print(labled)
```

```python
# adding a new column using a Series — index is aligned with the DataFrame index
salary_series = pd.Series([5000, 8000, 4500, 6000], index=["one","two","three","four"])
labled["salary"] = salary_series
print(labled)
```

```python
# deleting a column with del (modifies in place)
del labled["debt"]
print(labled)

# drop() returns a new DataFrame and does not modify the original
labled2 = labled.drop(columns=["salary"])
print(labled2)
```

```python
# converting a DataFrame to a numpy ndarray using .to_numpy()
arr = labled.to_numpy()
print(arr)
print(type(arr))
print(arr.dtype)  # dtype will be object when columns have mixed types
```
