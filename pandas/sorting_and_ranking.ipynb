```python
import pandas as pd
import numpy as np
s = pd.Series([10, 20, 30, 40], index=['d', 'c', 'b', 'a'])
# sort by the index,It sorts alphabetically by default in ascending order

s.sort_index()
# a    40
# b    30
# c    20
# d    10
# you can flip it by using ascending =false
s.sort_index(ascending=False)
# d    10
# c    20
# b    30
# a    40
```

```python
s = pd.Series([10, 20, 30, 40], index=['d', 'c', 'b', 'aa'])
# if you add a text as index or made type, pandas sorts strings lexicographically.
s.sort_index()
# aa    40   ← comes before b, c, d because 'a' < 'b'
# b     30
# c     20
# d     10
```

```python
data = pd.Series([10, 5, 3, 40])
# sort the values
print(data.sort_values())

# any missing values sorted to the end of series by default
data1 = pd.Series([10,2,np.nan,60])
print(data1.sort_values())
```

```python
obj = {'b':[4,7,-3,2],'a':[0,1,0,1]}
dataFrame= pd.DataFrame(obj)
print(dataFrame.sort_values(by='b'))
print(dataFrame.sort_values(by=['a','b']))

```

```python
s = pd.Series([7, 3, 5])

s.sort_values()   # reorders the rows
# 1    3
# 2    5
# 0    7

s.rank()          # keeps rows in place, adds rank numbers
# 0    3.0   ← 7 is the 3rd smallest
# 1    1.0   ← 3 is the 1st smallest
# 2    2.0   ← 5 is the 2nd smallest
```
