```python
import pandas as pd
# create normal series in pandas, the indecies by defualt is from 0 to n-1 as well known 
arr = pd.Series([10,2,12,45,10])
print(arr)
print(arr[0])

arr2 = pd.Series([3,5,7,12,11],index=["a","b","c","d","e"])
# you can specify the indecies in series 
print(arr2)
print(arr2["c"])
# you can get a list of indecies as series have two arrays on for values and the other for associated indecies
print(arr2.index)

# you can create a series by passing a dict to it 
data = {"a":10,"b":15,"d":45,"f":11}
arr3 = pd.Series(data)


print(arr3)
print(arr3.index)
```

```python
data = {"a":100,"b":15,"d":45,"f":11}
arr = pd.Series(data)
# you can make slicing at pandas series
print("3.Slicing : ",arr["a":"d"])

# positonal slicing
print("4.positional Slicing : ",arr.iloc[0:2])


 
```

```python
s = pd.Series([1, None, 3])
print(s.isna())  # return list of boolean items based on the item is none or not.
hasnNotNA=s.dropna() # return a new series with missing values removed
print(hasnNotNA)
filled=s.fillna(0) # replace nons with value 
print(filled)

```
