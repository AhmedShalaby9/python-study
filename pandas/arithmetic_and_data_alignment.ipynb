```python
import pandas as pd
import numpy as np

# data: 3x3, columns a/b/c, rows Cairo/Giza/Alex → values 0..8
data = pd.DataFrame(np.arange(9).reshape((3, 3)),
                    columns=['a', 'b', 'c'],
                    index=["Cairo", "Giza", "Alex"])

# data1: 4x3, columns b/d/e, rows Cairo/Giza/Alex/Banha → values 0..11
data1 = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     columns=['b', 'd', 'e'],
                     index=["Cairo", "Giza", "Alex", "Banha"])

# Adding aligns on both axes:
# - Columns a, c, d, e exist in only one DataFrame → NaN
# - Column b exists in both → summed for shared rows
# - Row Banha exists only in data1 → entire row is NaN
# - Result is sorted by index label (alphabetical)
print(data + data1)
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.arange(9).reshape((3, 3)),
                    columns=['a', 'b', 'c'],
                    index=["Cairo", "Giza", "Alex"])

data1 = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     columns=['b', 'd', 'e'],
                     index=["Cairo", "Giza", "Alex", "Banha"])

# fill_value=0 substitutes 0 wherever a label is missing in one DataFrame:
# - Columns a, c (only in data)  → treated as 0 in data1, so result = data value + 0
# - Columns d, e (only in data1) → treated as 0 in data,  so result = 0 + data1 value
# - Column b (in both)           → normal addition
# - Row Banha (only in data1)    → columns a, c are still NaN because data has NO Banha row at all
#   (fill_value only fills when the column exists in the other df; here a/c don't exist in data1)
result = data.add(data1, fill_value=0)

print(result)
```
