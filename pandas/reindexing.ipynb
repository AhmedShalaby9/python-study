```python
import pandas as pd

obj = pd.Series([10,4,78,44],index=["d","b","a","c"])
print(obj)
obj2 = obj.reindex(["a","b","c","d"])
# reindex function returns a new arrenged ojbect aligns with the new index
print(obj2)
obj3 = obj.reindex(["a","b","c","d","e"])
# introduce missing values if the index is not already in.
print(obj3)
```

```python
data = {"month":["October","April","May","June"],"age":[20,None,30,16],"job":["engineer","doctor","athletic","carpenter"]}
frame = pd.DataFrame(data,index=["a","c","b","d"])
# in data frames you can reindex using add new columns or remove the existing 
frame2 = frame.reindex(columns=["month","age"])

frame4 = frame.reindex(columns=["month","age","test"])

# reindex using index like series
frame3 = frame.reindex(index=["a","b","c","d"])

print(frame)
print(frame2)
print(frame3)
print(frame4)


```
