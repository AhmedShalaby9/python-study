```python
import pandas as pd

obj = pd.Series([10,4,78,44],index=["d","b","a","c"])
print(obj)
obj2 = obj.reindex(["a","b","c","d"])
# reindex function returns a new arranged object aligned with the new index
print(obj2)
obj3 = obj.reindex(["a","b","c","d","e"])
# introduce missing values if the index is not already in
print(obj3)
```

```python
data = {"month":["October","April","May","June"],"age":[20,None,30,16],"job":["engineer","doctor","athletic","carpenter"]}
frame = pd.DataFrame(data,index=["a","c","b","d"])
# in DataFrames you can reindex by columns or index
frame2 = frame.reindex(columns=["month","age"])
frame3 = frame.reindex(index=["a","b","c","d"])
frame4 = frame.reindex(columns=["month","age","test"])

print(frame)
print(frame2)
print(frame3)
print(frame4)
```

```python
# drop a label from a Series
obj4 = obj.drop("b")
print(obj4)
```

```python
# drop rows from a DataFrame by index label
frame5 = frame.drop(["a","c"])
print(frame5)
```

```python
# drop columns from a DataFrame using columns=
frame6 = frame.drop(columns=["job"])
print(frame6)
```

```python
# loc — label-based selection (uses index labels and column names)
print("original frame:")
print(frame)

print("\n--- single row ---")
print(frame.loc["a"])

print("\n--- multiple rows ---")
print(frame.loc[["a","b"]])

print("\n--- single value (row 'a', column 'job') ---")
print(frame.loc["a", "job"])

print("\n--- slice rows 'a' to 'b', columns 'month' to 'age' ---")
print(frame.loc["a":"b", "month":"age"])
```

```python
# iloc — position-based selection (uses integer positions, 0-indexed)
print(frame.iloc[0])                # first row
print(frame.iloc[[0, 2]])           # rows at position 0 and 2
print(frame.iloc[0, 2])             # value at row 0, column 2
print(frame.iloc[0:2, 0:2])         # slice first 2 rows and first 2 columns
```

```python
# boolean filtering with loc
print(frame.loc[frame["age"] > 18])           # rows where age > 18
print(frame.loc[frame["job"] == "doctor"])    # rows where job is doctor
```
