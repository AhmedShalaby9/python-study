```python
import pandas as pd
# Covariance — measures whether two columns move in the same direction and by how much  
df = pd.DataFrame({
    'study_hours': [1, 2, 3, 4, 5],
    'grade':       [50, 60, 70, 80, 90]
})
corr =df['study_hours'].cov(df['grade'])
# if the result is positive means both up together
# if the result is negative means one goes up and one goes down
# if the result is Zero or Near Zero means No Relation 
print(corr)
```

```python
# Correlation — covariance but normalized to always be between -1 and 1
corr =df['study_hours'].corr(df['grade'])
# 1 -> positive relation
# -1 -> negative relation
# 0 --> no relation
print(corr)
```

```python
obj = pd.Series(['c','d','a','c','b','a','f'])
uniques = obj.unique()
# unique func returns a series of the unique values
print(uniques)
# value counts returns the count of each item
counts = obj.value_counts()
print(counts)
# isin check the membership like is contain or not -- returns a boolean series
contains = obj.isin(['c','g'])
print(contains)
```

```python
# The core use case — aligning two datasets
# The main reason get_indexer exists is to figure out how to map one index onto another. For example, if you have two DataFrames with different row orders and you want to align them:

idx = pd.Index(['a', 'b', 'c', 'd'])

idx.get_indexer(['b', 'd', 'a'])
# array([1, 3, 0])  ← positions of b, d, a in the index
idx.get_indexer(['a', 'z', 'b'])
# array([ 0, -1,  1])
#              ↑ 'z' doesn't exist → -1


source = pd.Index(['a', 'b', 'c', 'd'])
target = pd.Index(['c', 'a', 'd'])

source.get_indexer(target)
# array([2, 0, 3])
# meaning: c is at position 2, a is at position 0, d is at position 3
```
